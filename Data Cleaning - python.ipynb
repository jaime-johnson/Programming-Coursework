```python
#Importing libraries
import pandas as pd
import numpy as np
```

```python
#Analysing data from 2008
year_2008 = pd.read_csv(r"C:\Users\jaime\Desktop\Coursework\2008.csv.bz2")
year_2008.head(-5)
```

```python
#Analysing data from 2007
year_2007= pd.read_csv(r"C:\Users\jaime\Desktop\Coursework\2007.csv.bz2")
year_2007.head(-5)
```

```python
#Analysing data from 2006
year_2006 = pd.read_csv(r"C:\Users\jaime\Desktop\Coursework\2006.csv.bz2")
year_2006.head(-5)
```

```python
#Merging data from 2006 and 2007
dataset = pd.concat([year_2006, year_2007])
```

```python
dataset.shape
```

```python
dataset.head(-5)
```

```python
#Checking the data types
dataset.dtypes
```

```python
#Checking for duplicated rows
dataset.duplicated().sum()
```

```python
#Removing duplicated rows
dataset = dataset.drop_duplicates()
```

```python
dataset.isnull().sum()
```

```python
#Dropping the CancellationCode column
dataset = dataset.drop(columns = 'CancellationCode')
```

```python
#Analysing the columns with null values further 
print(dataset["DepTime"].max())
print(dataset["ArrTime"].max())
```

```python
#Considering the DepTime and ArrTime only until midnight (2400hhmm)
dataset = dataset[(dataset["DepTime"]<2400) & (dataset["ArrTime"]<2400)]
```

```python
dataset.isnull().sum()
```

```python
#Creating a new column as TotalDelay by adding DepDelay and ArrDelay
dataset['TotalDelay'] = dataset['DepDelay'] + dataset['ArrDelay']
```

```python
dataset.shape
```

```python
file_path = 'cleaned_dataset_2006_2007.csv'
```

```python
dataset.to_csv(file_path, index=False)
```
