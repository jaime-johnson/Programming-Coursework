```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Defining the probability density function f(x)
def f(x):
    return 0.5*np.exp(-np.abs(x))
```

```python
#Given that
N=1000
s=1
```

```python
#Creating an array for x with all zeros initially
#The length of the array is N+1 because it has to store values from x_0 to x_N
x = np.zeros(N+1)
```

```python
# Step 1
x[0]= 1
```

```python
#Checking the length of the array x
len(x)
```

```python
#Step 2
for i in range(1,1001):
    #stimulating a random number x_star from normal distribution with mean x_i-1 and standard deviation s
    x_star = np.random.normal(x[i-1], s)
    #computing the ratio
    r = f(x_star) / f(x[i-1])
    #generating a random number u from uniform distribution between 0 and 1 
    u = np.random.uniform (0,1)
    #if u is less than r then x_i is equal to x_star or else x_i is equal to x_i-1
    if np.log(u) < np.log(r):
        x[i] = x_star
    else:
        x[i] = x[i-1]
```

```python
#Dropping the initial value from array so that the array x contains only the generated samples x_1 to x_N
x = x[1:]
```

```python
#Checking the length of the array
len(x)
```

```python
#Histogram
plt.hist(x, bins=35, density=True, color='lightgreen',edgecolor ='darkgreen', label='Histogram')
#Kernel Density plot
sns.kdeplot(x,color='black',label='Kernel Density Plot')
#graph of f(x)
values = np.linspace(-5,6,1500)
plt.plot(values,f(values),label='f(x)', color='orangered')


plt.xlabel("Generated Samples (x_1,...,x_N)")
plt.legend()
plt.savefig('Part 1(a)-python.png')
plt.show()
```

```python
#Calculating the mean of the generated sample
mean = np.mean(x)
print('Monte Carlo estimate of the mean is' , mean)
```

```python
#Calculating the standard deviation of the generated sample 
standard_deviation = np.std(x)
print('Monte Carlo estimate of the standard deviation is' , standard_deviation)
```
